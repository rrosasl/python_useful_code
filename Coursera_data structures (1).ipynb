# Dictionaries

Dictionaries are another type of collection, i..e having multiple values in one variables. 

A dictionary is a "bag"of values each with its own label. There's always a key and a value. A key it's a like a label. You can think of it as a baby database

Dictionaries use curly braces {}


```python
purse = dict()

purse['money'] = 12
purse['candy'] = 3
purse['tissues'] = 75
purse['list'] = [3,5,6]

print(purse)
print(purse['tissues'])

purse['candy'] += 8
print(purse)

type(purse)
```

```python
for key in purse:
    print(key)
```

Counting the number of names that appear

```python
counts = dict()
names = ['csev', 'cwen', 'csev','zqian', 'cwen']
for name in names:
    if name not in counts:
        counts[name] = 1
    else:
        counts[name] += 1
print(counts)
```

### Get() method

The patter of cheking to see if a key is already in a dictionary and assuming a default value if the key is not there is so common that there's a method called get() that does that

```python
counts = dict()
names = ['csev', 'cwen', 'csev','zqian', 'cwen']

for name in names:
    counts[name] = counts.get(name,0) + 1

print(counts)
```

### Counting patterns

```python
counts = dict()
line = 'Its Friday, Friday, Friday, got to get down on friday'
words = line.split()
print('Words:', words)

for word in words:
    counts[word] = counts.get(word,0) + 1

print(counts)
```

### Looping through dictionaries

Even though dictionaries are not stored in order, we can write a for loop that goes through all the entries. 

```python
counts = {'chuck':1, 'ricardo':10, 'jan': 20}
for key in counts:
    print(key,counts[key]) #by using dict[key] you can access the values
```

```python
counts = {'chuck':1, 'ricardo':10, 'jan': 20}
print(counts.values())
print(counts.items())
print(counts.keys())
```

You can even do 2 iteration variables!

```python
counts = {'chuck':1, 'ricardo':10, 'jan': 20}

for aaa,bbb in counts.items(): 
    #We usually can use for k, v as in for key and value
    print(aaa,bbb)
    
print(counts)
```

### Sorting lists of tuples
We can sort a list of tupes to get a sorted version of a dictionaary. First, we sort the dictionary by the keys using items() and sorted() function

```python
d = {'a':10,'c':1,'b':22}
print(sorted(d.items()))
print(d)
```

```python
#Sorting by key order
for k,v in sorted(d.items()):
    print(k,v)
```

```python
# Sorting by value order
c = {'a':10,'c':1,'b':22}
tmp = list()
for k,v in c.items():
    tmp.append( (v,k))
print('p1',tmp)

tmp = sorted(tmp, reverse=True)
print('p2', tmp)
tmp
```

Getting the top 10 most common words

```python
fhand = open('romeo.txt')
counts = dict()
for line in fhand: #Will iterate through every line
    words = line.split() #Will split the line into a list with all the words
    for word in words: #Will iterate through every word of the line
        counts[word] = counts.get(word,0) + 1  #Will append the word to the dictionary
        #If the word is not included it will have the value 0
counts
```

```python
lst = list()
for key, val in counts.items():
#Will iterate through the entire above dictionary
    newtup = val,key #THis new tuple will have first the value and then the key
    lst.append(newtup) # Will add the value and key to the new (before empty) list
lst = sorted(lst,reverse=True) #Will order from highes to lowest the list

for val, key in lst[:10]:
    print(key,val)
```

This second part could have been achieved by a single line of code

```python
print(sorted([(v,k) for k,v in counts.items()],reverse=True)[:10])
```

```python
fh = open('romeo.txt')
lst = list()

for line in fh:
    line = line.rstrip()
    sepa = line.split()
    for x in sepa:
        if x in lst : continue
        lst.append(x)
lst.sort()
lst
```

9.4 Write a program to read through the mbox-short.txt and figure out who has sent the greatest number of mail messages. The program looks for 'From ' lines and takes the second word of those lines as the person who sent the mail. The program creates a Python dictionary that maps the sender's mail address to a count of the number of times they appear in the file. After the dictionary is produced, the program reads through the dictionary using a maximum loop to find the most prolific committer.

```python
handle = open('mbox-short.txt')

counter = dict()

for line in handle:
    line = line.rstrip()
    if not line.startswith('From ') : continue
    line = line.split()[1]
    counter[line] = counter.get(line,0) + 1
print(counter)
```

```python
bigcount = None
bigword = None

for word, count in counter.items():
    if bigcount is None or count > bigcount: 
        bigword = word
        bigcount = count
    
print(bigword, bigcount)
```
